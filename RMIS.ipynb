### Read RMIS csv report


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
df = pd.read_csv("https://www.rmis.org/reports/CSV5048.txt")
with pd.option_context('display.max_rows', None, 'display.max_columns', None):  # more options can be specified also
    display(df)    
```

```python
df['first_release_date'] =  pd.to_datetime(df['first_release_date'], format='%Y%m%d')
df['last_release_date'] =  pd.to_datetime(df['last_release_date'], format='%Y%m%d')
df['total'] = (df['tagged_adclipped'] + df['tagged_unclipped'] + df['untagged_adclipped'] + 
               df['untagged_unclipped'] + df['untagged_unknown'])
df['unclipped'] = (df['tagged_unclipped'] + df['untagged_unclipped'] + df['untagged_unknown'])
df['prop_unclipped'] = df['unclipped']/df['total']
df['prop_untagged_unclipped'] = (df['untagged_unclipped']+ df['untagged_unknown'])/df['total']
```

```python
subset_df = df[(df['first_release_date'].dt.year >= 2005) & (df['first_release_date'].dt.year <= 2020)]
```

```python
with pd.option_context('display.max_rows', None, 'display.max_columns', None):  # more options can be specified also
    display(subset_df)   
```

```python
fig = plt.figure(figsize = (15,10))

plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%m/%d/%Y'))
plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval=365))
plt.scatter(subset_df['first_release_date'], subset_df['total'])

plt.gcf().autofmt_xdate()
```

```python
fig = plt.figure(figsize = (15,10))

plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%m/%d/%Y'))
# plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval=365))
plt.scatter(subset_df['first_release_date'], subset_df['untagged_unclipped'])
plt.ylabel("Number of untagged and unclipped", fontsize = 20)
plt.xlabel("Date", fontsize = 20)
plt.gcf().autofmt_xdate()
```

```python
fig = plt.figure(figsize = (15,10))

plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%m/%d/%Y'))
# plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval=365))
plt.scatter(subset_df['first_release_date'], subset_df['unclipped'])
plt.ylabel("Number of unclipped", fontsize = 20)
plt.xlabel("Date", fontsize = 20)
plt.gcf().autofmt_xdate()
```

```python
fig = plt.figure(figsize = (15,10))

plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%m/%d/%Y'))
# plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval=365))
plt.scatter(subset_df['first_release_date'], subset_df['prop_unclipped'])
plt.ylabel("Proportion of unclipped", fontsize = 20)
plt.xlabel("Date", fontsize = 20)
plt.gcf().autofmt_xdate()
```

```python
fig = plt.figure(figsize = (15,10))

plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%m/%d/%Y'))
# plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval=365))
plt.scatter(subset_df['first_release_date'], subset_df['prop_untagged_unclipped'], alpha = 0.5)
plt.ylabel("Proportion of untagged and unclipped", fontsize = 20)
plt.xlabel("Date", fontsize = 20)
plt.gcf().autofmt_xdate()
```
