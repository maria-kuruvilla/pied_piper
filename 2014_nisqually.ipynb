#### Changing 2014 Nisqually csv format to match other years 

```python
import numpy as np
import matplotlib.pyplot as plt
from sys import platform
import pandas as pd
import datetime
```

```python
df = pd.read_csv(data_string + "2014_Nisqually.csv")
```

```python
with pd.option_context('display.max_rows', None, 'display.max_columns', None):  # more options can be specified also
    display(df)
```

```python
for col in df.columns.values:
    print(col)
```

```python
#rename columns to be consistent with previous years
df_year = df.rename(columns={'StartDate': "Date",
                                  'EndTime': "Up_time",
                                  'StartTime': "Down_time",
                                  'EndDate': "Up_date",
                                  'Chinook/Fall/Yearling/NATURAL': "chinook1_wild_num",
                                  'Chinook/Fall/Zero/NATURAL': "chinook0_wild_num",
                                  'Chum/Fall/Zero/WILD': "chum0_wild_num",
                                  'Pink/NA/Zero/WILD': "pink0_wild_num",
                                  'Steelhead/Winter/Smolt - Unknown/WILD':'steelheadsmolt_wild_num',
                                  'Coho/NA/Smolt Yearling/NATURAL':'coho1_wild_num'})

for col in df_year.columns.values:
    print(col)
```

```python
df_year['Up_time'] = df_year['Up_time'].str.split(' ').str[1]
df_year['Down_time'] = df_year['Down_time'].str.split(' ').str[1]
```

```python
df_year['Down'] = pd.to_datetime(df_year['Date'] + ' ' + df_year['Down_time'])
df_year['Up'] = pd.to_datetime(df_year['Up_date'] + ' ' + df_year['Up_time'])
```

```python
df_year['midpoint'] = df_year['Down'] + (df_year['Up'] - df_year['Down'])/2

#get rid of columns with Trap status !=1 and fill 0 instead of NANs
df_year= df_year[df_year['TrapStatus'] == 1]
df_year.fillna(0, inplace=True)

with pd.option_context('display.max_rows', None, 'display.max_columns', None):  # more options can be specified also
    display(df_year)
```

```python
df_trial = pd.to_timedelta(df_year['Up'] - df_year['Down'], unit = 'hours')
```

```python
df_time = pd.to_timedelta(df_year['Up'] - df_year['Down']).dt.components[['days', 'hours', 'minutes', 'seconds']]
df_time['In'] = df_time['days']*24 + df_time['hours'] + df_time['minutes']/60
display(df_time)
```

```python
df_year['In'] = df_time['In']
df_year['chinook0_wild_perhour'] = df_year['chinook0_wild_num']/df_time['In']
df_year['chinook1_wild_perhour'] = df_year['chinook1_wild_num']/df_time['In']
df_year['coho1_wild_perhour'] = df_year['coho1_wild_num']/df_time['In']
df_year['chum0_wild_perhour'] = df_year['chum0_wild_num']/df_time['In']
df_year['pink0_wild_perhour'] = df_year['pink0_wild_num']/df_time['In']
df_year['steelheadsmolt_wild_perhour'] = df_year['steelheadsmolt_wild_num']/df_time['In']
df_final = df_year[['Date','Down_time','Up','Up_date','Up_time','In', 'chinook0_wild_num','chinook0_wild_perhour', 
                    'chinook1_wild_num','chinook1_wild_perhour', 'coho1_wild_num','coho1_wild_perhour', 
                    'steelheadsmolt_wild_num','steelheadsmolt_wild_perhour', 
                    'chum0_wild_num','chum0_wild_perhour',
                    'pink0_wild_num','pink0_wild_perhour']]
```

```python
df_final.to_csv(data_string+"2014_nisqually_all_R_new.csv")
```
