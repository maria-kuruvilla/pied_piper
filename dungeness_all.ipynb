```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import datetime
from sys import platform
```

```python

if platform == "linux" or platform == "linux2":
    data_string = "/home/maria/Documents/data/pied_piper/"

elif platform == "win32"or platform == "win64":
    data_string = "C:/Users/maria/OneDrive/Documents/data/pied_piper/"
```

```python
#2005

year = "2005"
csv_name = data_string + year + "_dungeness_all.csv"
df = pd.read_csv(csv_name, header=0)
with pd.option_context('display.max_rows', None, 'display.max_columns', None):  # more options can be specified also
    display(df)
```

```python
df['chinook0_wild_perhour'] = df['Chinook0_Wild']/df['In']

#chinook 1
df['chinook1_perhour'] = df['Chinook1_Total']/df['In']
df['chinook1_hatchery_perhour'] = (
    df['Chinook1_Hat']  )/df['In']
df['Chinook1_Wild'] = (df['Chinook1_Total'] - df['Chinook1_Hat'])
df['chinook1_wild_perhour'] = df['Chinook1_Wild']/df['In']

#coho 1
df['coho1_perhour'] = df['Coho1_Total']/df['In']
df['coho1_hatchery_perhour'] = (
    df['Coho1_Hat'] + df['Coho1_Hat1'])/df['In']
df['Coho1_Wild2'] = df['Coho1_Total'] - (df['Coho1_Hat'] + df['Coho1_Hat1'])
df['coho1_wild_perhour'] = (df['Coho1_Wild']+df['Coho1_Wild1']+df['Coho1_Wild2'])/df['In']

#steelhead
df['steelheadsmolt_perhour'] = df['Steelhead1_Total']/df['In']
df['steelheadsmolt_hatchery_perhour'] = (
    df['Steelhead1_Hat'])/df['In']
df['Steelhead1_Wild1'] = df['Steelhead1_Total'] - df['Steelhead1_Hat'] 
df['steelheadsmolt_wild_perhour'] = (
    df['Steelhead1_Wild']+df['Steelhead1_Wild1'])/df['In']
```

```python
with pd.option_context('display.max_rows', None, 'display.max_columns', None):  # more options can be specified also
    display(df)
```

```python
#function to plot fish per hour

def plot_fish_perhour(groupedby_dates_year, species = "chinook", year = "0", types=["hatchery", "wild", "total"]):
    
    fs = 18
    lw = 3
    transparency = 0.5
    plt.close('all') # always start by cleaning up
    fig = plt.figure(figsize=(12,8))
    ax1 = fig.add_subplot(111)
    for fish_type in types:

        if fish_type == "total":
            x_string = species + year + "_" + "perhour"
            fish_color = "gray"
        if fish_type == "wild":  
            x_string = species + year +"_" + "wild" + "_" + "perhour"
            fish_color = "salmon"
        if fish_type == "hatchery":  
            x_string = species + year +"_" + "hatchery" + "_" + "perhour"
            fish_color = "seagreen"
        if fish_type == "parr":  
            x_string = species + year +"_" + "parr" + "_" + "perhour"
            fish_color = "orangered"


        
        ax1.plot(
            groupedby_dates_year['Date'], groupedby_dates_year[x_string], color = fish_color,
            label = fish_type, alpha = 0.5, linewidth = lw)
    plt.gcf().autofmt_xdate()
    plt.xlabel('Date', size = 0.9*fs)
    plt.ylabel(species + " " +year + "+" + ' counts per hour', size = 0.9*fs)
    ax1.tick_params(labelsize=.8*fs)
    ax1.set_title(species + " " +year + "+" + ' counts', loc='left', fontsize = fs*1.5)
    plt.legend(fontsize=fs, loc='upper right', title = '')

    plt.show()
```

```python
#plot
plot_fish_perhour(df,"chinook","0",["wild","total","hatchery"])
```
