```python
#check if in-out for day is 24-hour or or no check or out,
#then change in_day to np.nan
#do same for in_night

# df['In_day'][df['In-Out for Day'] == 'no check']
# df['In_night'][df['In-Out for Day'] == 'no check']
# df['In_night'][df['In-Out for Day'] == '24-hr check']
# df['In_night'][df['In-Out for Day'] == 'OUT']
# df['In_night'][df['In-Out for Day'] == '24hr fish'] = np.nan
```

```python
def adjust_date_non24(row):
    if(row['In-Out for Day']!='24-hr check' and 
       row['In-Out for Day']!='24hr fish' and 
       row['In-Out for Day']!='no check' and
      row['Out_time_night'] > row['In_time_night']):
        return(row['Date'])
    else:
        return(row['Date']) + pd.DateOffset(days = 1)
    
df['Out_date_night'] = df.apply(adjust_date_non24,axis=1)
df['Out_datetime_night'] = pd.to_datetime(df['Out_date_night'].astype(str) + ' ' + df['Out_time_night'].astype(str) , 
                                        errors='coerce')
df['In_night'] = (df['Out_datetime_night'] - df['In_datetime_night']).astype('timedelta64[h]')
```

```python
#rename columns

df_rename = df.rename(columns={'Secchi Depth Day (cm)':'secchi_depth_day',
                               'Flow (cfs)':'flow',
                  '# of Fish Caught (Day)':'chinook0_wild_num_day',
                  'Daily CE (Inv. of SD model)':'trap_efficiency',
                   'Secchi Depth Night (cm)':'secchi_depth_night',
                  '# of Fish Caught (Night)':'chinook0_wild_num_night',
                   '# of Fish Caught (Day).1':'chinook0_hatchery_num_day',
                   '# of Fish Caught (Night).1':'chinook0_hatchery_num_night'
                  })

df_subset = df_rename[['Date','flow','secchi_depth_day','In_day','In_night', 'secchi_depth_night',
                'chinook0_wild_num_day',
                'chinook0_wild_num_night',
                'chinook0_hatchery_num_day',
                'chinook0_hatchery_num_night', 'trap_efficiency']]

df_subset['flow_day'] = df_subset['flow']
df_subset['flow_night'] = df_subset['flow']

df_subset['In'] = df_subset['In_day'].fillna(0) + df_subset['In_night'].fillna(0)

with pd.option_context('display.max_rows', None, 'display.max_columns', None):  # more options can be specified also
    display(df_subset)
```

```python
#if In day is nan, then chinook0_wild_num_day and chinook0_wild_num_day should be nan

df_subset.loc[(df_subset['In_day'] == np.nan),'chinook0_wild_num_day'] =np.nan

df_subset.loc[(df_subset['In_day'] == np.nan),'chinook0_hatchery_num_day'] =np.nan

df_subset.loc[(df_subset['In_night'] == np.nan),'chinook0_wild_num_night'] =np.nan

df_subset.loc[(df_subset['In_night'] == np.nan),'chinook0_hatchery_num_day'] =np.nan


df_subset['chinook0_wild_perhour_day'] = df_subset['chinook0_wild_num_day']/df_subset['In_day']
df_subset['chinook0_wild_perhour_night'] = df_subset['chinook0_wild_num_night']/df_subset['In_night']

df_subset['chinook0_hatchery_perhour_day'] = df_subset['chinook0_hatchery_num_day']/df_subset['In_day']
df_subset['chinook0_hatchery_perhour_night'] = df_subset['chinook0_hatchery_num_night']/df_subset['In_night']

df_subset['chinook0_wild_num_day_estimate'] = df_subset['chinook0_wild_num_day']/df_subset['trap_efficiency']
df_subset['chinook0_wild_num_night_estimate'] = df_subset['chinook0_wild_num_night']/df_subset['trap_efficiency']

df_subset['chinook0_wild_perhour_day_estimate'] = (df_subset['chinook0_wild_perhour_day']/
                                                   df_subset['trap_efficiency'])
df_subset['chinook0_wild_perhour_night_estimate'] = (df_subset['chinook0_wild_perhour_night']/
                                                   df_subset['trap_efficiency'])
df_subset['chinook0_wild_perhour_estimate'] = ((df_subset['chinook0_wild_num_day_estimate']+df_subset['chinook0_wild_num_night_estimate'])/
                                                   df_subset['In'])


df_subset['chinook0_hatchery_num_day_estimate'] = df_subset['chinook0_hatchery_num_day']/df_subset['trap_efficiency']
df_subset['chinook0_hatchery_num_night_estimate'] = df_subset['chinook0_hatchery_num_night']/df_subset['trap_efficiency']


df_subset['chinook0_hatchery_perhour_day_estimate'] = (df_subset['chinook0_hatchery_num_day_estimate']/
                                                   df_subset['In_day'])
df_subset['chinook0_hatchery_perhour_night_estimate'] = (df_subset['chinook0_hatchery_num_night_estimate']/
                                                   df_subset['In_night'])
df_subset['chinook0_hatchery_perhour_estimate'] = ((df_subset['chinook0_hatchery_num_day_estimate']+df_subset['chinook0_hatchery_num_night_estimate'])/
                                                   df_subset['In'])

with pd.option_context('display.max_rows', None, 'display.max_columns', None):  # more options can be specified also
    display(df_subset)
```
