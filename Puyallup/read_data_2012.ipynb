```python
# #read file in Windows
try:
    df = pd.read_excel(
   io=data_string + "puyallup_chinook_2004-2018.xlsx",
   sheet_name ='2012_edited',
    header = [4]
)
except PermissionError:
    print("Permission Error. Close the file, if it is open.")
    

df.head()
```

```python
pd.unique(df['In-Out_day']) #'-','OUT','out',nan
pd.unique(df['In-Out_night']) #'-','out',nan, 'Out'


df.columns
```

```python
df['In_day_edited'] = df['In-Out_day']
df['In_day_edited'][df['In-Out_day'] == 'Out'] = np.nan
df['In_day_edited'][df['In-Out_day'] == '-'] = np.nan
df['In_day_edited'][df['In-Out_day'] == 'out'] = np.nan
df['In_day_edited'][df['In-Out_day'] == 'OUT'] = np.nan
df['In_day_edited'].unique()
```

```python
df['In_night_edited'] = df['In-Out_night']
df['In_night_edited'][df['In-Out_night'] == 'Out'] = np.nan
df['In_night_edited'][df['In-Out_night'] == '-'] = np.nan
df['In_night_edited'][df['In-Out_night'] == 'out'] = np.nan
df['In_night_edited'][df['In-Out_night'] == 'OUT'] = np.nan
df['In_night_edited'].unique()
```

```python
df['In_time_day'] = pd.to_datetime(df['In_day_edited'].str.split('-',n=2, expand = True)[0], format = '%H%M').dt.time
df['Out_time_day'] = pd.to_datetime(df['In_day_edited'].str.split('-',n=2, expand = True)[1], format = '%H%M').dt.time
# df['Out_date_day'] = df['Date']+pd.Timedelta(days=1)
# df['Out_datetime_day'] = pd.to_datetime(df['Date'].astype(str) + ' ' + df['Out_time_day'].astype(str) , 
#                                         errors='coerce')
# df['In_datetime_day'] = pd.to_datetime(df['Date'].astype(str) + ' ' + df['In_time_day'].astype(str), 
#                                         errors='coerce')
# df['In_day'] = (df['Out_datetime_day'] - df['In_datetime_day']).astype('timedelta64[h]')
# df['In_day']

df['In_time_night'] = pd.to_datetime(df['In_night_edited'].str.split('-',n=2, expand = True)[0], format = '%H%M').dt.time
df['Out_time_night'] = pd.to_datetime(df['In_night_edited'].str.split('-',n=2, expand = True)[1], format = '%H%M').dt.time
# df['Out_date_night'] = df['Date']+pd.Timedelta(days=1)
# df['Out_datetime_night'] = pd.to_datetime(df['Out_date_night'].astype(str) + ' ' + df['Out_time_night'].astype(str) , 
#                                         errors='coerce')
df['In_datetime_night'] = pd.to_datetime(df['Date'].astype(str) + ' ' + df['In_time_night'].astype(str), 
                                        errors='coerce')

df['In_datetime_day'] = pd.to_datetime(df['Date'].astype(str) + ' ' + df['In_time_day'].astype(str), 
                                        errors='coerce')
```

```python
def adjust_date_night(row):
    if(row['Out_time_night'] > row['In_time_night']):
        return(row['Date'])
    else:
        return(row['Date'] + pd.DateOffset(days = 1))
    
def adjust_date_day(row):
    if(row['Out_time_day'] > row['In_time_day']):
        return(row['Date'])
    else:
        return(row['Date'] + pd.DateOffset(days = 1))
```

```python
df['Out_date_night'] = df.apply(adjust_date_night,axis=1)

df['Out_date_day'] = df.apply(adjust_date_day,axis=1)

df['Out_datetime_night'] = pd.to_datetime(df['Out_date_night'].astype(str) + ' ' + df['Out_time_night'].astype(str) , 
                                        errors='coerce')

df['Out_datetime_day'] = pd.to_datetime(df['Out_date_day'].astype(str) + ' ' + df['Out_time_day'].astype(str) , 
                                        errors='coerce')

df['In_night'] = (df['Out_datetime_night'] - df['In_datetime_night']).dt.total_seconds() / 3600

df['In_day'] = (df['Out_datetime_day'] - df['In_datetime_day']).dt.total_seconds() / 3600

```

```python
df.columns
```

```python
df_subset = df[['Date','flow','secchi_depth_day','In_day','In_night', 'secchi_depth_night',
                'chinook0_wild_num_day',
                       'chinook0_wild_num_day_estimate', 'chinook0_wild_num_night_estimate',
                'chinook0_wild_num_night',
                       'chinook0_wild_num',
                'chinook0_hatchery_num_day',
                'chinook0_hatchery_num_night', 'chinook0_hatchery_num','trap_efficiency_night']]

df_subset['flow_day'] = df_subset['flow']
df_subset['flow_night'] = df_subset['flow']

df_subset['In'] = df_subset['In_day'].fillna(0) + df_subset['In_night'].fillna(0)
df_subset['In'] = df_subset['In'].replace(0, np.nan)

df_subset.drop(df_subset.tail(2).index,inplace = True)
```

```python


#if In day is nan, then chinook0_wild_num_day and chinook0_wild_num_day should be nan

df_subset.loc[pd.isna(df_subset['In_day']),'chinook0_wild_num_day'] =np.nan
df_subset.loc[pd.isna(df_subset['In_day']),'chinook0_wild_num_day_estimate'] =np.nan

df_subset.loc[pd.isna(df_subset['In_day']),'chinook0_hatchery_num_day'] =np.nan

df_subset.loc[pd.isna(df_subset['In_night']),'chinook0_wild_num_night'] =np.nan
df_subset.loc[pd.isna(df_subset['In_night']),'chinook0_wild_num_night_estimate'] =np.nan

df_subset.loc[pd.isna(df_subset['In_night']),'chinook0_hatchery_num_day'] =np.nan


df_subset['trap_efficiency_day'] = df_subset['chinook0_wild_num_day']/df_subset['chinook0_wild_num_day_estimate'].replace(0,np.nan)

df_subset['chinook0_wild_perhour_day'] = df_subset['chinook0_wild_num_day']/df_subset['In_day']
df_subset['chinook0_wild_perhour_night'] = df_subset['chinook0_wild_num_night']/df_subset['In_night']

df_subset['chinook0_hatchery_perhour_day'] = df_subset['chinook0_hatchery_num_day']/df_subset['In_day']
df_subset['chinook0_hatchery_perhour_night'] = df_subset['chinook0_hatchery_num_night']/df_subset['In_night']

# df_subset['chinook0_wild_num_day_estimate'] = df_subset['chinook0_wild_num_day']/df_subset['trap_efficiency']
# df_subset['chinook0_wild_num_night_estimate'] = df_subset['chinook0_wild_num_night']/df_subset['trap_efficiency']

df_subset['chinook0_wild_perhour_day_estimate'] = (df_subset['chinook0_wild_perhour_day']/
                                                   df_subset['trap_efficiency_day'])
df_subset['chinook0_wild_perhour_night_estimate'] = (df_subset['chinook0_wild_perhour_night']/
                                                   df_subset['trap_efficiency_night'])


df_subset['chinook0_wild_perhour'] = df_subset['chinook0_wild_num']/df_subset['In']
df_subset['chinook0_hatchery_perhour'] = df_subset['chinook0_hatchery_num']/df_subset['In']

df_subset['chinook0_wild_perhour_estimate'] = (df_subset['chinook0_wild_perhour']/
                                                   df_subset['trap_efficiency_night'])



df_subset['chinook0_hatchery_num_day_estimate'] = df_subset['chinook0_hatchery_num_day']/df_subset['trap_efficiency_day']
df_subset['chinook0_hatchery_num_night_estimate'] = df_subset['chinook0_hatchery_num_night']/df_subset['trap_efficiency_night']


df_subset['chinook0_hatchery_perhour_day_estimate'] = (df_subset['chinook0_hatchery_num_day_estimate']/
                                                   df_subset['In_day'])
df_subset['chinook0_hatchery_perhour_night_estimate'] = (df_subset['chinook0_hatchery_num_night_estimate']/
                                                   df_subset['In_night'])
df_subset['chinook0_hatchery_perhour_estimate'] = (df_subset['chinook0_hatchery_perhour']/
                                                   df_subset['trap_efficiency_night'])
```

```python

df_subset['chinook0_wild_perhour_total'] = (df_subset['chinook0_wild_perhour_day'] + 
                                            df_subset['chinook0_wild_perhour_night'])

df_subset['chinook0_hatchery_perhour_total'] = (df_subset['chinook0_hatchery_perhour_day'] + 
                                            df_subset['chinook0_hatchery_perhour_night'])

df_subset['chinook0_wild_day_fraction'] = np.nan
df_subset['chinook0_hatchery_day_fraction'] = np.nan
n = len(df_subset['chinook0_wild_perhour_total'])

for i in range(n):
    if df_subset['chinook0_wild_perhour_day'].loc[i]==0 and df_subset['chinook0_wild_perhour_total'].loc[i] == 0:
        df_subset['chinook0_wild_day_fraction'].loc[i] = np.nan
        
    else:
        df_subset['chinook0_wild_day_fraction'].loc[i] = (df_subset['chinook0_wild_perhour_day'].loc[i]/
                                                          df_subset['chinook0_wild_perhour_total'].loc[i])
        
    if df_subset['chinook0_hatchery_perhour_day'].loc[i]==0 and df_subset['chinook0_hatchery_perhour_total'].loc[i] == 0:
        df_subset['chinook0_hatchery_day_fraction'].loc[i] = np.nan
        
    else:
        df_subset['chinook0_hatchery_day_fraction'].loc[i] = (df_subset['chinook0_hatchery_perhour_day'].loc[i]/
                                                          df_subset['chinook0_hatchery_perhour_total'].loc[i])
        
df_subset = df_subset.drop(['chinook0_wild_perhour_total','chinook0_hatchery_perhour_total'],axis = 1)
```
