```python

df_temp = df[['Date','temp']]
df_temp['temp'] = pd.to_numeric(df_temp["temp"], errors="coerce")
# df_temp.head()
df_temp['temp'].unique()
df_temp['temp'].isnull().sum()
```

```python
#merge temperature data with puyallup data

df_temp['Date'] = pd.to_datetime(df_temp['Date'])
df_temp['daynight_category'] = df_temp['Date'].dt.hour.between(6, 18, inclusive=True) \
    .replace({True: 'day', False: 'night'})
df_temp['datetime'] = df_temp['Date']
df_temp['Date'] = df_temp['Date'].dt.date
df_temp.head(50)
```

```python
df_temp_agg = df_temp.groupby(['Date','daynight_category']).agg({'temp':'mean'}).reset_index()
# df_temp.pivot(index = "Date",columns="daynight_category", values="temp")
df_temp_agg.to_csv(data_string + 'white_temp_date.csv')

```
