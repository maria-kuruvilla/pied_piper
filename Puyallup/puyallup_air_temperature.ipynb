```python
temp = pd.read_csv(data_string + 'air_temp.csv', header=0)
temp.columns = ['Date', 'temp_F']
temp['temp'] = (temp['temp_F'] - 32)*5/9
temp['Date'] = pd.to_datetime(temp['Date'])
temp['temp'] = temp['temp'].interpolate('linear')
temp
```

```python
filter1 = (temp.Date.dt.month == 12) & (temp.Date.dt.day >= 21)
temp['temp'] = temp['temp'].round(1)
temp['water_year'] = temp.Date.dt.year.where(~(filter1),temp.Date.dt.year+1)

temp['atu_solstice'] = round(temp.groupby(['water_year'])['temp'].cumsum(),1)

with pd.option_context('display.max_rows', None, 'display.max_columns', None):  # more options can be specified also
    display(temp)
```
