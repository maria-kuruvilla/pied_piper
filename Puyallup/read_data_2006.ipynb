```python
# #read file in Windows
try:
    df = pd.read_excel(
   io=data_string + "puyallup_chinook_2004-2018.xlsx",
   sheet_name ='2006_edited',
    header = [2]
)
except PermissionError:
    print("Permission Error. Close the file, if it is open.")
    

df.head()
```

```python
list(df)

df_rename = df.rename(columns = {
    'Flow (cfs)':'flow',
    'Secchi Depth Day (cm)':'secchi_depth',
    'Hours of Day':'In_day',
    '# of Fish Caught (Day)':'chinook0_wild_num_day',
    'CPUE (Day)':'chinook0_wild_perhour_day',
    'Estimated Production Day':'chinook0_wild_day_num_estimate',
    
    'Hours of Night':'In_night',
    
    '# of Fish Caught (Night)': 'chinook0_wild_num_night',
    'CPUE (Night)':'chinook0_wild_perhour_night',
    'Estimated Production Night':'chinook0_wild_night_num_estimate',
    
    'Total CPUE':'chinook0_wild_perhour',
    'Total fish captured':'chinook0_wild_num',
    'Daily Production Estimates':'chinook0_wild_num_estimate',
    
    'Total Hours':'In',
    
    '# of Fish Caught (Day).1':'chinook0_hatchery_num_day',
    'CPUE (Day).1':'chinook0_hatchery_perhour_day',
    'Estimated Production Day.1':'chinook0_hatchery_day_num_estimate',
    
    'Hours of Night':'In_night',
    
    '# of Fish Caught (Night).1':'chinook0_hatchery_num_night',
    'CPUE (Night).1':'chinook0_hatchery_perhour_night',
    'Estimated Production Night.1':'chinook0_hatchery_night_num_estimate',
    
    'Total CPUE.1':'chinook0_hatchery_perhour',
    'Total Fish Captured':'chinook0_hatchery_num',
    'Daily Production Estimates.1':'chinook0_hatchery_num_estimate'
    
}, inplace = False)

df_rename.head()
```

```python
#note to self - need flow and secchi depth even on days the trap was not in!!!!

df_subset = df_rename[['Date', 'flow', 'secchi_depth','In_day','In_night','In',
                'chinook0_wild_num_day','chinook0_wild_perhour_day',
                'chinook0_wild_num_night','chinook0_wild_perhour_night',
                'chinook0_hatchery_num_day','chinook0_hatchery_perhour_day',
                'chinook0_hatchery_num_night','chinook0_hatchery_perhour_night',
                'chinook0_wild_perhour','chinook0_hatchery_perhour',
                'chinook0_wild_day_num_estimate','chinook0_wild_night_num_estimate',
               'chinook0_hatchery_day_num_estimate','chinook0_hatchery_night_num_estimate']]

df_subset.loc[(df_subset['In_day'] == '-'),'In_day':'chinook0_hatchery_night_num_estimate'] =np.nan
df_subset.loc[(df_subset['In_day'] == 'Out'),'In_day':'chinook0_hatchery_night_num_estimate'] =np.nan
df_subset.loc[(df_subset['In_day'] == 'Out '),'In_day':'chinook0_hatchery_night_num_estimate'] =np.nan
df_subset.loc[(df_subset['In_day'] == 'OUT'),'In_day':'chinook0_hatchery_night_num_estimate'] =np.nan
df_subset.loc[(df_subset['In_day'] == 'ST'),'In_day':'chinook0_hatchery_night_num_estimate'] =np.nan

df_subset.loc[(df_subset['In_night'] == '-'),'In_day':'chinook0_hatchery_night_num_estimate'] =np.nan
df_subset.loc[(df_subset['In_night'] == 'ST'),'In_day':'chinook0_hatchery_night_num_estimate'] =np.nan
df_subset.loc[(df_subset['In_night'] == 'Out'),'In_day':'chinook0_hatchery_night_num_estimate'] =np.nan
df_subset.loc[(df_subset['In_night'] == 'Out '),'In_day':'chinook0_hatchery_night_num_estimate'] =np.nan
df_subset.loc[(df_subset['In_night'] == 'OUT'),'In_day':'chinook0_hatchery_night_num_estimate'] =np.nan

df_subset['chinook0_wild_perhour_day_estimate'] = (df_subset['chinook0_wild_day_num_estimate']/
                                                   df_subset['In_day'])
df_subset['chinook0_wild_perhour_night_estimate'] = (df_subset['chinook0_wild_night_num_estimate']/
                                                   df_subset['In_night'])
df_subset['chinook0_wild_perhour_estimate'] = ((df_subset['chinook0_wild_day_num_estimate']+df_subset['chinook0_wild_night_num_estimate'])/
                                                   df_subset['In'])

df_subset['chinook0_hatchery_perhour_day_estimate'] = (df_subset['chinook0_hatchery_day_num_estimate']/
                                                   df_subset['In_day'])
df_subset['chinook0_hatchery_perhour_night_estimate'] = (df_subset['chinook0_hatchery_night_num_estimate']/
                                                   df_subset['In_night'])
df_subset['chinook0_hatchery_perhour_estimate'] = ((df_subset['chinook0_hatchery_day_num_estimate']+df_subset['chinook0_hatchery_night_num_estimate'])/
                                                   df_subset['In'])

with pd.option_context('display.max_rows', None, 'display.max_columns', None):  # more options can be specified also
    display(df_subset)
```

```python
df_subset['chinook0_wild_perhour_total'] = (df_subset['chinook0_wild_perhour_day'] + 
                                            df_subset['chinook0_wild_perhour_night'])

df_subset['chinook0_hatchery_perhour_total'] = (df_subset['chinook0_hatchery_perhour_day'] + 
                                            df_subset['chinook0_hatchery_perhour_night'])

df_subset['chinook0_wild_day_fraction'] = np.nan
df_subset['chinook0_hatchery_day_fraction'] = np.nan
n = len(df_subset['chinook0_wild_perhour_total'])

for i in range(n):
#     print(i)
    if df_subset['chinook0_wild_perhour_day'].loc[i]==0 and df_subset['chinook0_wild_perhour_total'].loc[i] == 0:
        df_subset['chinook0_wild_day_fraction'].loc[i] = np.nan
    
    elif df_subset['chinook0_wild_perhour_day'].loc[i]==np.nan and df_subset['chinook0_wild_perhour_total'].loc[i] == np.nan:
        df_subset['chinook0_wild_day_fraction'].loc[i] = np.nan
        
    elif df_subset['chinook0_wild_perhour_day'].loc[i]=='-' or df_subset['chinook0_wild_perhour_total'].loc[i] == '-':
        df_subset['chinook0_wild_day_fraction'].loc[i] = np.nan
        
    else:
        df_subset['chinook0_wild_day_fraction'].loc[i] = (df_subset['chinook0_wild_perhour_day'].loc[i]/
                                                          df_subset['chinook0_wild_perhour_total'].loc[i])
        
    if df_subset['chinook0_hatchery_perhour_day'].loc[i]==0 and df_subset['chinook0_hatchery_perhour_total'].loc[i] == 0:
        df_subset['chinook0_hatchery_day_fraction'].loc[i] = np.nan
        
    elif df_subset['chinook0_hatchery_perhour_day'].loc[i]==np.nan and df_subset['chinook0_hatchery_perhour_total'].loc[i] == np.nan:
        df_subset['chinook0_hatchery_day_fraction'].loc[i] = np.nan
        
    elif df_subset['chinook0_hatchery_perhour_day'].loc[i]=='-' or df_subset['chinook0_hatchery_perhour_total'].loc[i] == '-':
        df_subset['chinook0_hatchery_day_fraction'].loc[i] = np.nan
        
    else:
        df_subset['chinook0_hatchery_day_fraction'].loc[i] = (df_subset['chinook0_hatchery_perhour_day'].loc[i]/
                                                          df_subset['chinook0_hatchery_perhour_total'].loc[i])
        
with pd.option_context('display.max_rows', None, 'display.max_columns', None):  # more options can be specified also
    display(df_subset)
```

```python
#save the data

df_subset.to_csv(data_string + "2006_puyallup.csv")  
```
