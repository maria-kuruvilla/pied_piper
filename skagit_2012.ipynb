```python

try:
    df_2012 = pd.read_excel(
   io= data_string + "2012 Skagit mainstem catch.xls",
   sheet_name ='Screw',
    header = [1,2]
)
except PermissionError:
    print("Permission Error. Close the file, if it is open.")
    

df_2012.head()
columns = list(df_2012)
print(columns)

df_2012.columns  = ['_'.join(col) for col in df_2012.columns.values]
print(df_2012.columns)

df_2012 = df_2012.rename(columns={'Unnamed: 0_level_0_Page #': "Page",
                                  'Unnamed: 1_level_0_Date Dn': "Date",
                                  'Unnamed: 2_level_0_Time Dn': "Down_time",
                                  'Unnamed: 4_level_0_Time Up': "Up_time",
                                 'Unnamed: 3_level_0_Date Up': "Up_date"})

df_2012 = df_2012.iloc[:-21]

df_2012['Up_time'] = pd.to_datetime(df_2012['Up_time'], format='%H%M').dt.time
df_2012['Down_time'] = pd.to_datetime(df_2012['Down_time'], format='%H%M').dt.time



df_2012['Up'] = pd.to_datetime((df_2012['Up_date'].astype(str)) + ' ' + (df_2012["Up_time"].astype(str)))
df_2012['Down'] = pd.to_datetime((df_2012['Date'].astype(str)) + ' ' + (df_2012["Down_time"].astype(str)))
df_2012['midpoint'] = df_2012['Down'] + (df_2012['Up'] - df_2012['Down'])/2
df_2012['In'] = df_2012['Up'] - df_2012['Down']

df_2012['In'] = df_2012['In']/ np.timedelta64(1, 'h')

with pd.option_context('display.max_rows', None, 'display.max_columns', None):  # more options can be specified also
    display(df_2012)
```

```python
df_2012.columns = df_2012.columns.str.strip()
print(df_2012.columns)
```

```python
df_year = df_2012.rename(columns={'Wild Chin 0+_Total': "chinook0_wild_num",
                                 'Hatch Chin 0+_Total': "chinook0_hatchery_num",
                                 'Wild Chin 1+_Total': "chinook1_wild_num",
                                 'Hatch Chin 1+_Total': "chinook1_hatchery_num",
                                  'Coho 0+_Total':'coho0_wild_total',
                                  'Wild Coho 1+_Total':'coho1_wild_num',
                                  'Hatch Coho 1+_Total': "coho1_hatchery_num",
                                  'Wild Sthd_Total':'steelheadsmolt_wild_num',
                                  'Pink_Total': 'pink0_wild_num',
                                 'Chum_Total': 'chum0_wild_num'})

df_year_subset = df_year[['Date', 'Down_time', 'Up_date', 'Up_time', 'midpoint', 'In', 'Up', 'Down',
                         'chinook0_hatchery_num', 'chinook0_wild_num',
                         'chinook1_hatchery_num', 'chinook1_wild_num',
                         'coho1_hatchery_num', 'coho1_wild_num',
                         'chum0_wild_num', 'pink0_wild_num','steelheadsmolt_wild_num']]

#convet nans to 0s
df_year_subset.fillna(0, inplace=True)
df_year_subset['chinook0_wild_num'].iloc[194] = 0 #it was showing up as black for some reason. filling it with 0.


df_year_subset['chinook0_hatchery_perhour'] = df_year_subset['chinook0_hatchery_num']/df_year_subset['In']
df_year_subset['chinook0_wild_perhour'] = df_year_subset['chinook0_wild_num']/df_year_subset['In']

df_year_subset['chinook1_hatchery_perhour'] = df_year_subset['chinook1_hatchery_num']/df_year_subset['In']
df_year_subset['chinook1_wild_perhour'] = df_year_subset['chinook1_wild_num']/df_year_subset['In']


df_year_subset['coho1_hatchery_perhour'] = df_year_subset['coho1_hatchery_num']/df_year_subset['In']
df_year_subset['coho1_wild_perhour'] = df_year_subset['coho1_wild_num']/df_year_subset['In']

df_year_subset['chum0_wild_perhour'] = df_year_subset['chum0_wild_num']/df_year_subset['In']
df_year_subset['pink0_wild_perhour'] = df_year_subset['pink0_wild_num']/df_year_subset['In']
df_year_subset['steelheadsmolt_wild_perhour'] = df_year_subset['steelheadsmolt_wild_num']/df_year_subset['In']




with pd.option_context('display.max_rows', None, 'display.max_columns', None):  # more options can be specified also
    display(df_year_subset)
```

```python
#covariates

url = 'https://waterdata.usgs.gov/nwis/dv?cb_00010=on&cb_00060=on&cb_00065=on&format=rdb&site_no=12178000&legacy=&referred_module=sw&period=&begin_date=2011-12-21&end_date=2012-12-21'

html = urlopen(url).read()
soup = BeautifulSoup(html, features="html.parser")

# kill all script and style elements
for script in soup(["script", "style"]):
    script.extract()    # rip it out

# get text
text = soup.get_text()
text_data = text[1907:]
df = pd.read_csv(io.StringIO(text_data), sep = '\t', header = None, names = ["USGS", "random0", "Date", "flow", "random2","temp_max", "random3", 
                                                                       "temp_min", "random4","temp", "random5"])

df_temp_flow_subset = df[['Date','flow','temp']]
df_temp_flow_subset['Date'] = pd.to_datetime(df_temp_flow_subset['Date'])
df_temp_flow_subset = df_temp_flow_subset.rename(columns={'temp_mean':'temp',})
with pd.option_context('display.max_rows', None, 'display.max_columns', None):  # more options can be specified also
    display(df_temp_flow_subset)
    
df_temp_flow_subset['atu_solstice'] = df_temp_flow_subset['temp'].cumsum()

year = df_temp_flow_subset['Date'].dt.year[0]
atu_df_april = df_temp_flow_subset.loc[(df_temp_flow_subset['Date'] >= str(year+1)+'-4-01')]
atu_df_april['atu_april'] = atu_df_april['temp'].cumsum()
atu_df_april

df_atu = df_temp_flow_subset.merge(atu_df_april[['Date','atu_april']], how = 'left', on='Date')
df_atu['flow_diff'] = df_atu['flow'].diff()
df_atu['temp_diff'] = df_atu['temp'].diff()
df_atu['doy'] = df_atu['Date'].dt.dayofyear
with pd.option_context('display.max_rows', None, 'display.max_columns', None):  # more options can be specified also
    display(df_atu)
```

```python
df_photo = pd.read_csv(data_string + 'photoperiod_skagit.csv')
df_photo = df_photo[['doy','photoperiod']]
df_atu_photo = df_atu.merge(df_photo, how = 'left', on='doy')

df_lunar = pd.DataFrame(pd.date_range(start='2011-12-21', end='2012-12-21'), columns = ['Date'])
df_lunar.head()
moon_list = []
for i in df_lunar['Date']:
    a=ephem.Moon(i)
    mn=a.moon_phase #illuminated percentage of the moon.
    moon_list.append(mn)

df_lunar['lunar_phase'] = moon_list
df_lunar.head()
df_atu_photo_lunar = df_atu_photo.merge(df_lunar, how = 'left', on='Date')
df_atu_photo_lunar['photo_diff'] = df_atu_photo_lunar['photoperiod'].diff()
with pd.option_context('display.max_rows', None, 'display.max_columns', None):  # more options can be specified also
    display(df_atu_photo_lunar)
```

```python
df_screw_covariates = df_year_subset.merge(df_atu_photo_lunar, how = 'left', on='Date')
with pd.option_context('display.max_rows', None, 'display.max_columns', None):  # more options can be specified also
    display(df_screw_covariates)
```
