```python
#read the original scoop and screw 2010 covariates without all the covariates (only temp and flow)

df_screw_orig = pd.read_csv(data_string + '2010_skagit_screw.csv')
df_scoop_orig = pd.read_csv(data_string + '2010_skagit_scoop.csv')
df_screw_orig.head()
# df_scoop_orig.head()
```

```python
df_screw_orig['Date'] = pd.to_datetime(df_screw_orig['Date'])
datetimes = df_screw_orig['Date'].tolist() # these are existing datetimes - will add here the missing
dates = [x.date() for x in datetimes] # these are converted to dates

min_date = min(dates)
max_date = max(dates)
for d in range((max_date - min_date).days):
    forward_date = min_date + datetime.timedelta(d)
    if forward_date not in dates:
        datetimes.insert(d, np.datetime64(forward_date))
trial = pd.DataFrame({'Date': datetimes})

df_screw_orig = trial.merge(df_screw_orig, on=['Date'], how='left')
```

```python
df_scoop_orig['Date'] = pd.to_datetime(df_scoop_orig['Date'])
datetimes = df_scoop_orig['Date'].tolist() # these are existing datetimes - will add here the missing
dates = [x.date() for x in datetimes] # these are converted to dates

min_date = min(dates)
max_date = max(dates)
for d in range((max_date - min_date).days):
    forward_date = min_date + datetime.timedelta(d)
    if forward_date not in dates:
        datetimes.insert(d, np.datetime64(forward_date))
trial = pd.DataFrame({'Date': datetimes})

df_scoop_orig = trial.merge(df_scoop_orig, on=['Date'], how='left')
```
