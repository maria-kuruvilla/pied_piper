Goal - To read scoop, screw, day, night files of each night, aggregate by day to get one aggregated file

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import datetime
import matplotlib.dates as mdates
```

```python
from sys import platform
if platform == "linux" or platform == "linux2":
    data_string = "/home/maria/Documents/data/pied_piper/skagit/"
    output_string = "/home/maria/Documents/output/pied_piper/skagit/"

elif platform == "win32"or platform == "win64":
    data_string = "C:/Users/maria/OneDrive/Documents/data/pied_piper/skagit/"
    output_string = "C:/Users/maria/OneDrive/Documents/output/pied_piper/skagit/"
```

```python
year = 2013
traps = ['scoop','screw']
day_night = ['day','night']
csv_names = []
for trap in traps:
    for time in day_night:
        csv_names.append(data_string+str(year)+"_skagit_" +trap +"_covariates_"  + time + ".csv")
    
df = pd.concat([pd.read_csv(csv_name, header=0) for csv_name in csv_names])

with pd.option_context('display.max_rows', None, 'display.max_columns', None):  # more options can be specified also
    display(df)
```

```python
df_agg = df.groupby('Date').agg({'In' : 'sum', 
                        'chinook0_wild_num' : 'sum',
                       'chinook0_hatchery_num' : 'sum',
                       'chinook1_wild_num' : 'sum',
                       'chinook1_hatchery_num' : 'sum',
                       'chum0_wild_num' : 'sum',
                       'coho0_wild_num' : 'sum',
                       'coho1_wild_num' : 'sum',
                       'coho1_hatchery_num' : 'sum',
                       'pink0_wild_num' : 'sum',
                       'sockeye0_wild_num' : 'sum',
                       'sockeye1_wild_num' : 'sum',
                       'steelheadsmolt_wild_num' : 'sum',
                       'flow' : 'mean',
                       'temp' : 'mean',
                       'atu_solstice' : 'mean',
                       'atu_april' : 'mean',
                       'photoperiod' : 'mean',
                       'lunar_phase' : 'mean',})

with pd.option_context('display.max_rows', None, 'display.max_columns', None):  # more options can be specified also
    display(df_agg)
```

```python
df_agg['Date'] = pd.to_datetime(df_agg.index)
df_agg['doy'] = df_agg['Date'].dt.dayofyear
df_agg['photo_diff'] = np.diff(df_agg['photoperiod'], prepend = np.nan)
df_agg['temp_diff'] = np.diff(df_agg['temp'], prepend = np.nan)
df_agg['flow_diff'] = np.diff(df_agg['flow'], prepend = np.nan)

df_agg['chinook0_wild_perhour'] = df_agg['chinook0_wild_num']/df_agg['In']
df_agg['chinook0_hatchery_perhour'] = df_agg['chinook0_hatchery_num']/df_agg['In']

df_agg['chinook1_wild_perhour'] = df_agg['chinook1_wild_num']/df_agg['In']
df_agg['chinook1_hatchery_perhour'] = df_agg['chinook1_hatchery_num']/df_agg['In']

df_agg['chum0_wild_perhour'] = df_agg['chinook0_wild_num']/df_agg['In']
df_agg['coho0_wild_perhour'] = df_agg['chinook0_wild_num']/df_agg['In']

df_agg['coho1_wild_perhour'] = df_agg['coho1_wild_num']/df_agg['In']
df_agg['coho1_hatchery_perhour'] = df_agg['coho1_hatchery_num']/df_agg['In']

df_agg['pink0_wild_perhour'] = df_agg['pink0_wild_num']/df_agg['In']
df_agg['sockeye0_wild_perhour'] = df_agg['sockeye0_wild_num']/df_agg['In']
df_agg['sockeye1_wild_perhour'] = df_agg['sockeye1_wild_num']/df_agg['In']
df_agg['steelheadsmolt_wild_perhour'] = df_agg['steelheadsmolt_wild_num']/df_agg['In']

with pd.option_context('display.max_rows', None, 'display.max_columns', None):  # more options can be specified also
    display(df_agg)
```

```python
df_agg.to_csv(data_string + str(year) + '_skagit.csv', index=True)
```
