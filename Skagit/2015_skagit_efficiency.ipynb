```python
# Define the ACCDB file path
accdb_file_path = data_string + '2015 Skagit Mainstem.accdb'

# Define the query for 'qry_DateSummaryRelease'
query = 'SELECT * FROM qry_DateSummaryRelease'

# Create a connection string
conn_str = (
    r'DRIVER={Microsoft Access Driver (*.mdb, *.accdb)};'
    r'DBQ=' + accdb_file_path + ';'
)

# Establish a connection
conn = pyodbc.connect(conn_str)

# Execute the query and load the result into a pandas DataFrame
df_release = pd.read_sql_query(query, conn)

# Close the connection
conn.close()

# Display the DataFrame
df_release.head()
```

```python
# Define the ACCDB file path
accdb_file_path = data_string + '2015 Skagit Mainstem.accdb'

# Define the query for 'qry_Recoveries'
query = 'SELECT * FROM qry_Recoveries'

# Create a connection string
conn_str = (
    r'DRIVER={Microsoft Access Driver (*.mdb, *.accdb)};'
    r'DBQ=' + accdb_file_path + ';'
)

# Establish a connection
conn = pyodbc.connect(conn_str)

# Execute the query and load the result into a pandas DataFrame
df_recoveries = pd.read_sql_query(query, conn)

# Close the connection
conn.close()

# Display the DataFrame
df_recoveries.head()
```

```python
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    display(df_recoveries)

with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    display(df_release)
```

```python
# Drop rows with '5' and '13' in 'SpeciesRunLifeOrigin_Id'
df_recoveries = df_recoveries[~df_recoveries['SpeciesRunLifeOrigin_Id'].isin([5, 13])]


with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    display(df_recoveries)
```

```python
# Drop rows with 'chum' in 'SPECIES_Desc'
df_release = df_release[df_release['SPECIES_Desc'] != 'Chum']


# Convert 'ReleaseDate' to datetime type
df_release['ReleaseDate'] = pd.to_datetime(df_release['ReleaseDate'])


# Group by 'ReleaseDate' and sum the 'SumOfNumReleased'
df_release['SumOfNumReleased'] = df_release.groupby('ReleaseDate')['SumOfNumReleased'].transform('sum')


# Drop duplicates based on 'ReleaseDate'
df_release = df_release.drop_duplicates(subset='ReleaseDate')


with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    display(df_release)
```

```python
# Convert date columns to datetime
df_release['ReleaseDate'] = pd.to_datetime(df_release['ReleaseDate'])
df_recoveries['EndDate'] = pd.to_datetime(df_recoveries['EndDate'])


# Offset the date by one day
df_release['ReleaseDate'] += pd.DateOffset(1)


# Merge based on the common date columns
df_2015 = pd.merge(df_recoveries, df_release[['ReleaseDate', 'SumOfNumReleased']], left_on = 'EndDate', right_on = 'ReleaseDate', how = 'left')


# Drop the duplicate date column
df_2015 = df_2015.drop('ReleaseDate', axis = 1)


with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    display(df_2015)
```

```python
# Calculate the trap efficiency
    # For scoop
df_2015['trap_efficiency_scoop'] = df_2015['Skagit River Scoop'] / df_2015['SumOfNumReleased']
    # For screw
df_2015['trap_efficiency_screw'] = df_2015['Skagit River Screw'] / df_2015['SumOfNumReleased']


# Insert missing dates from the beginning of the year and use the same trap efficiency values of the first row
data = []
data.insert(0, {'EndDate': pd.to_datetime('2015-01-01'), 'trap_efficiency_scoop': 0.014267,
                'trap_efficiency_screw': 0.010375})
df_2015 = pd.concat([pd.DataFrame(data), df_2015], ignore_index=True)


with pd.option_context('display.max_rows', None, 'display.max_columns', None):
     display(df_2015)
```

```python
# Create new columns for the start date and end date based of the date ranges in the 'Date.1' column
df_2015['start_date']  = df_2015['EndDate']
df_2015['end_date'] = df_2015['EndDate'].shift(-1)
df_2015['end_date'] = df_2015['end_date'].fillna(pd.to_datetime('2015-12-31'))


# with pd.option_context('display.max_rows', None, 'display.max_columns', None):
#      display(df_2015)
```

```python
# Add a new column 'row' with row numbers
df_2015['row'] = range(1, len(df_2015) + 1)


# Combine the trap effeciencies for scoop & screw into 1 column
df_2015_long = pd.wide_to_long(df_2015, stubnames='trap_efficiency', i='row',j='trap',sep='_', suffix=r'\w+')


with pd.option_context('display.max_rows', None, 'display.max_columns', None):
     display(df_2015_long)
```

```python
# Reset the index of the dataframe
df_2015_long = df_2015_long.reset_index()
# df_2015_long
```

```python
# Save as a csv file
df_2015_long.to_csv(output_string + 'skagit_2015_long.csv', index = True)
```
