#### All files have chum and pink num and per hour for wild and hatchery

```python
def aggregate(year):
    csv_name = data_string+str(year)+"_dungeness_all.csv"
    df_year_flow = pd.read_csv(csv_name, header=0)
#     if year == 2015:
#         df_year_flow['temp'] = np.nan
    groupedby_dates_year = df_year_flow.groupby(by=["Date"]).agg({'In':'sum',
                                                                  'chinook0_hatchery_perhour':'sum',
                                                                  'chinook0_wild_perhour':'sum',
                                                                  'chinook0_hatchery_num':'sum',
                                                                  'chinook0_wild_num':'sum',
                                                                  'chinook1_hatchery_perhour':'sum',
                                                                  'chinook1_wild_perhour':'sum',
                                                                  'chinook1_hatchery_num':'sum',
                                                                  'chinook1_wild_num':'sum',
                                                                  'coho1_hatchery_perhour':'sum',
                                                                  'coho1_wild_perhour':'sum',
                                                                  'coho1_hatchery_num':'sum',
                                                                  'coho1_wild_num':'sum',
                                                                  'steelheadsmolt_hatchery_perhour':'sum',
                                                                  'steelheadsmolt_wild_perhour':'sum',
                                                                  'steelheadsmolt_hatchery_num':'sum',
                                                                  'steelheadsmolt_wild_num':'sum',
                                                                  'pink0_hatchery_perhour':'sum',
                                                                  'pink0_wild_perhour':'sum',
                                                                  'pink0_hatchery_num':'sum',
                                                                  'pink0_wild_num':'sum',
                                                                  'chum0_hatchery_perhour':'sum',
                                                                  'chum0_wild_perhour':'sum',
                                                                  'chum0_hatchery_num':'sum',
                                                                  'chum0_wild_num':'sum',
                                                                  'flow':'mean',
                                                                  'temp':'mean'})
    groupedby_dates_year.to_csv(data_string+str(year)+"_dungeness.csv")
```

```python
years = range(2006,2021)
for year in years:
    print(year)
    aggregate(year)
```
