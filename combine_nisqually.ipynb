```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import datetime
```

```python
from sys import platform
if platform == "linux" or platform == "linux2":
    data_string = "/home/maria/Documents/data/pied_piper/nisqually/"
    output_string = "/home/maria/Documents/output/pied_piper/nisqually/"

elif platform == "win32"or platform == "win64":
    data_string = "C:/Users/maria/OneDrive/Documents/data/pied_piper/nisqually/"
    output_string = "C:/Users/maria/OneDrive/Documents/output/pied_piper/nisqually/"
```

```python
years = range(2014,2021)
csv_names = []
for year in years:
    csv_names.append(data_string+str(year)+"_nisqually_all_R_new.csv")
    
df = pd.concat([pd.read_csv(csv_name, header=0) for csv_name in csv_names])

#day of year
df['Date'] = pd.to_datetime(df['Date'])
df['doy'] = df['Date'].dt.dayofyear
df['year'] = df['Date'].dt.year
with pd.option_context('display.max_rows', None, 'display.max_columns', None):  # more options can be specified also
    display(df)                                             
```

```python
df.to_csv(data_string + "nisqually_2014-2020.csv")  
```

```python
flow_data = pd.read_csv(data_string + "flow_data.csv")
flow_data
```

```python
#combine flow data with nisqually trap data

flow_data_subset = flow_data[['Date','flow']]
flow_data_subset['Date'] = pd.to_datetime(flow_data_subset['Date'])
df_year_flow = pd.merge_asof(df.dropna(subset = ['Date']),flow_data_subset, on="Date",tolerance=pd.Timedelta(days = 1))
df_year_flow
```

```python
plt.plot(df_year_flow['Date'], df_year_flow['flow'])
```

```python
df_year_flow.to_csv(data_string + "nisqually_2014-2020.csv")  
```
