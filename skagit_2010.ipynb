```python

try:
    df_2010 = pd.read_excel(
   io= data_string + "2010 Skagit mainstem catch.xls",
   sheet_name ='Screw',
    header = [1,2]
)
except PermissionError:
    print("Permission Error. Close the file, if it is open.")
    

df_2010.head()
columns = list(df_2010)
print(columns)

df_2010.columns  = ['_'.join(col) for col in df_2010.columns.values]
print(df_2010.columns)

df_2010 = df_2010.rename(columns={'Unnamed: 0_level_0_Page #': "Page",
                                  'Unnamed: 1_level_0_Date Dn': "Date",
                                  'Unnamed: 2_level_0_Time Dn': "Down_time",
                                  'Unnamed: 4_level_0_Time Up': "Up_time",
                                 'Unnamed: 3_level_0_Date Up': "Up_date",
                                 'Unnamed: 5_level_0_Stat Week':'Week'})

with pd.option_context('display.max_rows', None, 'display.max_columns', None):  # more options can be specified also
    display(df_2010)
```

```python
df_year_subset = df_year[['Date', 'Down_time', 'Up_date', 'Up_time', 'midpoint', 'In', 'Up', 'Down',
                         'chinook0_hatchery_num', 'chinook0_wild_num',
                         'chinook1_hatchery_num', 'chinook1_wild_num',
                         'coho1_hatchery_num', 'coho1_wild_num',
                         'chum0_wild_num', 'pink0_wild_num','steelheadsmolt_wild_num']]
with pd.option_context('display.max_rows', None, 'display.max_columns', None):  # more options can be specified also
    display(df_year_subset)
```

```python
#Let's look at the scoop trap


try:
    df_2010 = pd.read_excel(
   io= data_string + "2010 Skagit mainstem catch.xls",
   sheet_name ='Scoop',
   skipinitialspace = True,
    header = [1,2]
)
except PermissionError:
    print("Permission Error. Close the file, if it is open.")
    

df_2010.head()
columns = list(df_2010)
print(columns)

df_2010.columns  = ['_'.join(col) for col in df_2010.columns.values]
df_2010.columns = df_2010.columns.str.strip()
print(df_2010.columns)
```

```python
df_year_scoop = df_2010.rename(columns={'Unnamed: 0_level_0_Page #':'Page', 'Unnamed: 1_level_0_Date Dn':'Date',
       'Unnamed: 2_level_0_Time Dn':"Down_time", 'Unnamed: 3_level_0_Date Up':'Up_date',
       'Unnamed: 4_level_0_Time Up':'Up_time', 'Unnamed: 5_level_0_Stat Week':'Week',
       'Wild Chin 0+_Total':'chinook0_wild_num', 'Hatch Chin 0+_Total':'chinook0_hatchery_num',
       'Wild Chin 1+_Total':'chinook1_wild_num', 'Hatch Chin 1+_Total':'chinook1_hatchery_num',
       'Chum_Total':'chum0_wild_num',
       'Coho 0+_Total':'coho0_wild_num',
       'Wild Coho 1+_Total':'coho1_wild_num', 'Hatch Coho 1+_Total':'coho1_hatchery_num',
       'Pink_Total':'pink0_wild_num', 'Wild Sthd_Total':'steelheadsmolt_wild_num'})
```

```python
df_year_scoop = df_year_scoop[:-2]
df_year_scoop['Up_time'] = pd.to_datetime(df_year_scoop['Up_time'], format='%H%M').dt.time
df_year_scoop['Down_time'] = pd.to_datetime(df_year_scoop['Down_time'], format='%H%M').dt.time
```

```python
#convet nans to 0s
df_year_scoop.fillna(0, inplace=True)
df_year_scoop['Up'] = pd.to_datetime((df_year_scoop['Up_date'].astype(str)) + ' ' + (df_year_scoop["Up_time"].astype(str)))
df_year_scoop['Down'] = pd.to_datetime((df_year_scoop['Date'].astype(str)) + ' ' + (df_year_scoop["Down_time"].astype(str)))
df_year_scoop['midpoint'] = df_year_scoop['Down'] + (df_year_scoop['Up'] - df_year_scoop['Down'])/2
df_year_scoop['In'] = df_year_scoop['Up'] - df_year_scoop['Down']
df_year_scoop['In'] = df_year_scoop['In']/ np.timedelta64(1, 'h')
df_year_scoop_subset = df_year_scoop[['Date', 'Down_time', 'Up_date', 'Up_time', 'midpoint', 'In', 'Up', 'Down',
                         'chinook0_hatchery_num', 'chinook0_wild_num',
                         'chinook1_hatchery_num', 'chinook1_wild_num',
                         'coho1_hatchery_num', 'coho1_wild_num',
                         'chum0_wild_num', 'pink0_wild_num','steelheadsmolt_wild_num']]
with pd.option_context('display.max_rows', None, 'display.max_columns', None):  # more options can be specified also
    display(df_year_scoop_subset)
```

```python
# url_flow = 'https://waterdata.usgs.gov/nwis/dv?cb_00010=on&cb_00060=on&cb_00065=on&format=html&site_no=12178000&legacy=&referred_module=sw&period=&begin_date=2010-01-01&end_date=2010-12-16'


#getting temperature data
#reading html data
year = '2010'
url = 'https://waterdata.usgs.gov/nwis/dv?cb_00010=on&cb_00060=on&cb_00065=on&format=rdb&site_no=12178000&legacy=&referred_module=sw&period=&begin_date=2010-01-01&end_date=2010-12-16'

html = urlopen(url).read()
soup = BeautifulSoup(html, features="html.parser")

# kill all script and style elements
for script in soup(["script", "style"]):
    script.extract()    # rip it out

# get text
text = soup.get_text()
text_data = text[1945:]
df = pd.read_csv(io.StringIO(text_data), sep = '\t', header = None, names = ["USGS", "random0", "Date", "flow", "random1",
                                                                       "flow_height", "random2","temp_max", "random3", 
                                                                       "temp_min", "random4","temp", "random5"])

df_temp_flow_subset = df[['Date','flow','temp']]
df_temp_flow_subset['Date'] = pd.to_datetime(df_temp_flow_subset['Date'])
# df_temp_flow_subset = df_temp_flow_subset.rename(columns={'temp_mean':'temp',})
with pd.option_context('display.max_rows', None, 'display.max_columns', None):  # more options can be specified also
    display(df_temp_flow_subset)
```

```python
########### next steps 

#add flow and tem data to screw data

df_screw_merge = df_year_subset.merge(df_temp_flow_subset, on = 'Date', how = 'left', validate = 'many_to_one')

# with pd.option_context('display.max_rows', None, 'display.max_columns', None):  # more options can be specified also
#     display(df_screw_merge)

df_screw_merge.to_csv(data_string+'2010_skagit_screw_all.csv')


```

```python
# add flow and temp to scoop data

# save those data files


df_scoop_merge = df_year_scoop_subset.merge(df_temp_flow_subset, on = 'Date', how = 'left', validate = 'many_to_one')

# with pd.option_context('display.max_rows', None, 'display.max_columns', None):  # more options can be specified also
#     display(df_scoop_merge)

df_scoop_merge.to_csv(data_string+'2010_skagit_scoop_all.csv')


```

```python
#group screw data by date and save file
df_screw_merge_grouped = df_screw_merge.groupby(by = ['Date']).agg({'In':'sum',
                                                             'chinook0_hatchery_num':'sum', 
                                                             'chinook0_wild_num':'sum',
                                                             'chinook1_hatchery_num':'sum', 
                                                             'chinook1_wild_num':'sum',
                                                             'coho1_hatchery_num':'sum', 
                                                             'coho1_wild_num':'sum',
                                                             'chum0_wild_num':'sum', 
                                                             'pink0_wild_num':'sum',
                                                             'steelheadsmolt_wild_num':'sum',
                                                             'temp':'mean',
                                                             'flow':'mean'})

# with pd.option_context('display.max_rows', None, 'display.max_columns', None):  # more options can be specified also
#     display(df_screw_merge_grouped)

df_screw_merge_grouped.to_csv(data_string+'2010_skagit_screw.csv')


```

```python
#group scoop data by date and save file
df_scoop_merge_grouped = df_scoop_merge.groupby(by = ['Date']).agg({'In':'sum',
                                                             'chinook0_hatchery_num':'sum', 
                                                             'chinook0_wild_num':'sum',
                                                             'chinook1_hatchery_num':'sum', 
                                                             'chinook1_wild_num':'sum',
                                                             'coho1_hatchery_num':'sum', 
                                                             'coho1_wild_num':'sum',
                                                             'chum0_wild_num':'sum', 
                                                             'pink0_wild_num':'sum',
                                                             'steelheadsmolt_wild_num':'sum',
                                                             'temp':'mean',
                                                             'flow':'mean'})


df_scoop_merge_grouped.to_csv(data_string+'2010_skagit_scoop.csv')



# with pd.option_context('display.max_rows', None, 'display.max_columns', None):  # more options can be specified also
#     display(df_scoop_merge_grouped)
```

```python
#combine scoop and screw and save the file

#group by date and save that file


df_scoop_screw_merge = pd.concat([df_screw_merge,df_scoop_merge])

df_scoop_screw_merge_grouped = df_scoop_screw_merge.groupby(by = ['Date']).agg({'In':'sum',
                                                             'chinook0_hatchery_num':'sum', 
                                                             'chinook0_wild_num':'sum',
                                                             'chinook1_hatchery_num':'sum', 
                                                             'chinook1_wild_num':'sum',
                                                             'coho1_hatchery_num':'sum', 
                                                             'coho1_wild_num':'sum',
                                                             'chum0_wild_num':'sum', 
                                                             'pink0_wild_num':'sum',
                                                             'steelheadsmolt_wild_num':'sum',
                                                             'temp':'mean',
                                                             'flow':'mean'})

df_scoop_screw_merge_grouped.to_csv(data_string+'2010_skagit_scoop_screw.csv')

df_scoop_screw_merge.to_csv(data_string+'2010_skagit_scoop_screw_all.csv')
```
